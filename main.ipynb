```python
import logging as lg

lg.basicConfig(filename="logs/test.log",level=lg.INFO,format="%(asctime)s:%(levelname)s:%(message)s")

lg.info("This is an info message")
lg.warning("This is a warning message")
```

```python
def div(a,b):
    try:
        lg.info(str(a)+"/"+str(b))
        d = a/b
        return d
    except Exception as e:
        print("check log")
        lg.error("division by zero error")
        lg.exception(str(e))
        lg.critical("critical error")

div(4,0)
```
